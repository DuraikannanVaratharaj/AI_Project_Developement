```python
sentences = [
    ["python", "developer", "job"],
    ["machine", "learning", "engineer"],
    ["data", "science", "project"],
    ["backend", "developer", "python"]
]
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences,
    vector_size=100,
    window=5,
    min_count=1,
    workers=4,
    sg=1   # Skip-gram
)

model.save("word2vec.model")
```

```python
print(model.wv["python"])  # vector
```

```python
print(model.wv.most_similar("python"))
```

```python
model.wv.save_word2vec_format("word2vec.txt")
```
